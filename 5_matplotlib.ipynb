## Basic graphs. Format.

Matplotlib is the most widely used python library for plotting data. It has a wide number of plot types, in 2 and 3 dimensions and allows to customize many details of the plots.

We will first show the general syntax with simple examples, and show how to display the graphs in a jupyter notebook and/or save them to files.

There are different ways to create a graph using matplotlib. This notebook shows one particular method, one that is very flexible and thus makes it possible to refine the results more. The method is based on the use of the function `subplots` from the `pyplot` submodule. 

### Matplotlib and pyplot

`pyplot` is a submodule of matplotlib, the one that we use in most cases to generate the graphs. It is a common practice to import pyplot with the following line:

```python
import matplotlib.pyplot as plt
```

Thereafter we will use `plt.` to gain access to many functions and methods inside pyplot.

### Graphs using `plt.subplots`

With matplotlib we always plot ___data___, which means that we need a list, array, series of numbers to plot.

In this example, we will use the following lists of numbers to generate plots:

```python
list1 = [ 1, 2, 4, 5, 7, 8, 9 ]
list2 = [ 0, 3, 3, 5, 6, 8, 6 ]
list3 = [ 0, 3, 4, 0, 1, 3, 2 ]
```

```python
fig, ax = plt.subplots() # Create a figure and axis using plt.subplots

ax.plot( list1 ) # Plot some data in the axis

plt.show( fig ) # Show the figure in the screen

plt.close( fig ) # Close the figure
```

```python
fig, ax = plt.subplots() # Create a figure and axis using plt.subplots

ax.plot( list2 ) # Plot some data in the axis

plt.show( fig ) # Show the figure in the screen

plt.close( fig ) # Close the figure
```

```python
fig, ax = plt.subplots() # Create a figure and axis using plt.subplots

ax.plot( list3 ) # Plot some data in the axis

plt.show( fig ) # Show the figure in the screen

plt.close( fig ) # Close the figure
```

It is useful to visualize the previous piece of code as composed of three parts:

```python
#---------------------------------------------------------------------
# Part 1: Create a figure and axis using plt.subplots
fig, ax = plt.subplots()
#---------------------------------------------------------------------

# Part 2: Plot, modify, plot again, include text, etc.
ax.plot( list3 )

#---------------------------------------------------------------------
# Part 3: Show, save to file, and close
plt.show( fig ) 
plt.close( fig ) 
#---------------------------------------------------------------------
```

The most work will be on part 2. The first thing that we can test is to plot all three lists of data on the same axis:

```python
#---------------------------------------------------------------------
# Part 1: Create a figure and axis using plt.subplots
fig, ax = plt.subplots()
#---------------------------------------------------------------------

# Part 2: Plot, modify, plot again, include text, etc.
ax.plot( list1 )
ax.plot( list2 )
ax.plot( list3 )

#---------------------------------------------------------------------
# Part 3: Show, save to file, and close
plt.show( fig ) 
plt.close( fig ) 
#---------------------------------------------------------------------
```

Note that each call to `ax.plot()` line adds a new series to the graph.

Also notice that each element in the series is posotioned in the position that it has in the array. For example, with `list1`:

```python
print( list1 )
```

This happens because there is no concrete specification of the position of the values on the x axis, and therefore `matplotlib` puts them one after the other, in ther position on the list.

We can specify the position on the `x` axis by sending two values to `.plot()`, first the `x` positions and then the `y` values:

If we use the values on `list1` as `x` coordinates, we get the following graph:

```python
list1
```

```python
list2
```

```python
#---------------------------------------------------------------------
# Part 1: Create a figure and axis using plt.subplots
fig, ax = plt.subplots()
#---------------------------------------------------------------------

# Part 2: Plot, modify, plot again, include text, etc.
ax.plot( list1, list2 )
ax.plot( list1, list3 )

#---------------------------------------------------------------------
# Part 3: Show, save to file, and close
plt.show( fig ) 
plt.close( fig ) 
#---------------------------------------------------------------------
```

In other words, when we send two series of values to `ax.plot()`, each value in the first one defines where the corresponding value in the second will be positioned.

### Format

We will modify the same graphic to show some ways to make it look better, namely:
    
* change the line's width, style and color
* add markers for the data points
* add a legend
* set labels for both axis, and a title

#### Line width, color and style

Each call to `ax.plot` plots a series of data. We can add parameters to that series to format it, just after the data parameters. 

Some commonly used format parameters are:
* `linewidth` → takes numeric values, default is 1
* `linestyle` → defines for example dashed lines, takes values like: '-', '--', '-.'
* `color` → it is possible to specify colors by name ('red', 'black'), or abbreviations

For more detailed information check `ax.plot?`

```python
plt.plot?
```

```python
#---------------------------------------------------------------------
# Part 1: Create a figure and axis using plt.subplots
fig, ax = plt.subplots()
#---------------------------------------------------------------------

# Part 2: Plot, modify, plot again, include text, etc.
ax.plot( list1, list2, linewidth=5, linestyle=':', color='black' )
ax.plot( list1, list3, linewidth=2, linestyle='--', color='red', marker='o' )

ax.text( 4.5, 7, "Maximum point →", fontsize=12)

#---------------------------------------------------------------------
# Part 3: Show, save to file, and close
plt.show( fig ) 
plt.close( fig ) 
#---------------------------------------------------------------------
```

#### Markers

We can add markers to each point in the data series. A marker can be a circle, a square or many other symbols that tell us where each data point is located. They can be combined with lines, or used insted of them.

```python
#---------------------------------------------------------------------
# Part 1: Create a figure and axis using plt.subplots
fig, ax = plt.subplots()
#---------------------------------------------------------------------

# Part 2: Plot, modify, plot again, include text, etc.
ax.plot( list1, list2, marker='o', markersize=20, markerfacecolor='white', markeredgecolor='red' ) #circle markers
ax.plot( list1, list3, marker='s' ) # square markers

#---------------------------------------------------------------------
# Part 3: Show, save to file, and close
plt.show( fig ) 
plt.close( fig ) 
#---------------------------------------------------------------------
```

Common options used for formatting markers are:
* `markersize`
* `markerfacecolor`
* `markeredgecolor`

You can combine the use of lines and markers, and also plot the same data twice to get different colors:

```python
#---------------------------------------------------------------------
# Part 1: Create a figure and axis using plt.subplots
fig, ax = plt.subplots()
#---------------------------------------------------------------------

# Part 2: Plot, modify, plot again, include text, etc.
ax.plot( list1, list2, linewidth=1, color='black' )
ax.plot( list1, list2, linewidth=0, marker='o', markersize=7, color='green' ) # circle markers
ax.plot( list1, list3, linewidth=0, marker='s' ) # square markers

#---------------------------------------------------------------------
# Part 3: Show, save to file, and close
plt.show( fig ) 
plt.close( fig ) 
#---------------------------------------------------------------------
```

#### Legend

For each series of data in a plot, we can define a `label`, which we later include in a legend. Using the last graph, we can then define labels and include a legend as follows:

```python
#---------------------------------------------------------------------
# Part 1: Create a figure and axis using plt.subplots
fig, ax = plt.subplots( figsize=(15,7) )
#---------------------------------------------------------------------

# Part 2: Plot, modify, plot again, include text, etc.
ax.plot( list1, list2, linewidth=1, color='black', label='list2-black' )
ax.plot( list1, list2, linewidth=0, marker='o', markersize=17, color='green', label='list2-green' )
ax.plot( list1, list3, linewidth=0, marker='s', label='list3' )

ax.legend( loc='best', title="Measuring sites 2020", fontsize=15 ) # Takes the 'labels' from each data series and displays a legend
#---------------------------------------------------------------------
# Part 3: Show, save to file, and close
plt.show( fig ) 
plt.close( fig ) 
#---------------------------------------------------------------------
```

Note that there are two entries for the list2 data. This comes from the two lines that plot that data. We can just define an empty label to those data that we don't want to be in the legend:

```python
#---------------------------------------------------------------------
# Part 1: Create a figure and axis using plt.subplots
fig, ax = plt.subplots()
#---------------------------------------------------------------------

# Part 2: Plot, modify, plot again, include text, etc.
ax.plot( list1, list2, 
         linewidth=1, color='black',
         marker='o', markersize=7, markerfacecolor='green', 
         label='list2-green' )

ax.plot( list1, list3, linewidth=0, marker='s', label='list3' )

ax.legend()
#---------------------------------------------------------------------
# Part 3: Show, save to file, and close
plt.show( fig ) 
plt.close( fig ) 
#---------------------------------------------------------------------
```

It is possible to move the legend to a fixed location using the following coordinates. Otherwise, it will be placed in the _best_ place, which means somewhere where there is no plot, or at least not very much.

For example, to manually place the legend in the center, use: `ax.legend( loc='center' )`

```python
#---------------------------------------------------------------------
# Part 1: Create a figure and axis using plt.subplots
fig, ax = plt.subplots()
#---------------------------------------------------------------------

# Part 2: Plot, modify, plot again, include text, etc.
ax.plot( list1, list2, linewidth=1, color='black', label='' )
ax.plot( list1, list2, linewidth=0, marker='o', markersize=7, color='green', label='list2-green' )
ax.plot( list1, list3, linewidth=0, marker='s', label='list3' )

ax.legend( loc='center', title='Two series of data' )
#---------------------------------------------------------------------
# Part 3: Show, save to file, and close
plt.show( fig ) 
plt.close( fig ) 
#---------------------------------------------------------------------
```

#### Axis labels and figure title

To add labels to the `x` and `y` axes we use: `ax.set_xlabel()` and `ax.set_xlabel()`.
    
We can use formulas within the text with the "dollar-syntax": \\$2x+3/a\\$ → $2x+3/a$

And we use `fontsize` to set, well, the font size.


```python
#---------------------------------------------------------------------
# Part 1: Create a figure and axis using plt.subplots
fig, ax = plt.subplots()
#---------------------------------------------------------------------

# Part 2: Plot, modify, plot again, include text, etc.

# Plot
ax.plot( list1, list2, linewidth=1, color='black', label='' )
ax.plot( list1, list2, linewidth=0, marker='o', markersize=7, color='green', label='list2-green' )
ax.plot( list1, list3, linewidth=0, marker='s', label='list3' )

# Format
ax.set_xlabel( 'Time $[s]$', fontsize=14 )
ax.set_ylabel( 'Acceleration $[ \mu m \cdot s^{-2}]$', fontsize=14 )
ax.legend( loc='best', title='Two series of data' )
ax.grid( ) # Add grid on both axes
fig.suptitle( 'Sup-title of the complete figure', fontsize=18 )

#---------------------------------------------------------------------
# Part 3: Show, save to file, and close
plt.show( fig ) 
plt.close( fig ) 
#---------------------------------------------------------------------
```

### Predefined styles

`matplotlib` includes a number of formatting styles that can be used easily to change the overall looking of the graphs.

A complete list of these styles can be consulted using the following command:

```python
plt.style.available
```

You can use these styles in two forms:

* activate the style "permanently", and use it for the rest of the session or until you change it explicitly again
* make only one single graph with a style and change again to the one that was previously active.

We can use this simple graph to test different styles. Having it wrapped in a function makes it easy to test.

```python
def simple_graph():
    #---------------------------------------------------------------------
    # Part 1: Create a figure and axis using plt.subplots
    fig, ax = plt.subplots()
    #---------------------------------------------------------------------

    # Part 2: Plot, modify, plot again, include text, etc.

    # Plot
    ax.plot( list1, list2, linewidth=1, color='black', label='' )
    ax.plot( list1, list2, linewidth=0, marker='o', markersize=7, color='green', label='list2-green' )
    ax.plot( list1, list3, linewidth=0, marker='s', label='list3' )

    # Format
    ax.set_xlabel( 'Time $[s]$', fontsize=14 )
    ax.set_ylabel( 'Acceleration $[ m / s^2]$', fontsize=14 )
    ax.legend( loc='best', title='Two series of data' )
    ax.grid( ) # Add grid on both axes
    fig.suptitle( 'Sup-title of the complete figure', fontsize=18 )

    #---------------------------------------------------------------------
    # Part 3: Show, save to file, and close
    plt.show( fig ) 
    plt.close( fig ) 
    #---------------------------------------------------------------------
```

```python
simple_graph()
```

To activate a style "permanently" use the following command:

```python
plt.style.use( 'ggplot' ) # Set 'ggplot' as the plotting style from now on
```

### Freshweight: Bar diagram

We will use the data from the lettuce experiment to create a simple bar diagram showing the average weight of the three treatments.

For that, we use the `pandas` functionality to import the data, and also to create tables grouped by treatment.

```python
import pandas as pd
```

We read the excel file 'lettuce.xlsx', and we need the data in the 'Freshweight' sheet:

```python
df_freshweight = pd.read_excel( 'data/lettuce.xlsx', sheet_name='Freshweight' )
df_freshweight.head()
```

Now, we group by treatment, using the mean as aggregation function. There are three treatments, we should therefore obtain 3 average results.

```python
means = df_freshweight.groupby( by='Treatment' ).mean()
```

The function `groupby` will return the mean of all columns in the `df_freshweight` dataframe:

```python
means
```

To create a bar diagram we use the same syntax from the previous diagrams, but we replace the function `ax.plot` by `ax.bar`:

```python
#---------------------------------------------------------------------
# Part 1: Create a figure and axis using plt.subplots
fig, ax = plt.subplots()
#---------------------------------------------------------------------

# Part 2: Plot, modify, plot again, include text, etc.
individual_bars = ax.bar( means.index, means['Freshweight (g)'], width=0.4 )

colors = [ 'blue', 'yellow', 'black' ]

for i, bar in enumerate(individual_bars):
    individual_bars[i].set_color(colors[i])
    
#---------------------------------------------------------------------
# Part 3: Show, save to file, and close
plt.show( fig ) 
plt.close( fig ) 
#---------------------------------------------------------------------
```

A second element that is very easy to include in this diagram are error bars. There are error bars in x and in y. We need bars in the y-axis, and use therefore the `yerr` argument of the function.

We need to pass this argument in either of two forms:
* as a list of values, one for each bar: [ val1, val2, val3 ] → these will be the length of the error bars above and below the bars (this is the one we want to include with the corresponding standard deviation)
* as a list of lists, 2 values for each bar: [ [ val1_below, val1_above ], [ val2_below, val2_above ], [ val3_below, val3_above ] ] → this would be needed to include maximum and minimum values, for example

To calculate the standard deviation we use the same syntax that with the means:

```python
stds = df_freshweight.groupby( by='Treatment' ).std()
```

And we include that line in the `yerr` parameter. Remember to specify the column that we use: 'Freshweight (g)'.

```python
stds['Freshweight (g)']
```

```python
#---------------------------------------------------------------------
# Part 1: Create a figure and axis using plt.subplots
fig, ax = plt.subplots()
#---------------------------------------------------------------------

manual_errors = [ 50, 100, 200 ]

# Part 2: Plot, modify, plot again, include text, etc.
ax.bar( means.index, means['Freshweight (g)'], width=0.4, yerr=manual_errors )

#---------------------------------------------------------------------
# Part 3: Show, save to file, and close
plt.show( fig ) 
plt.close( fig ) 
#---------------------------------------------------------------------
```

Lastly, we can supperpose a plot to the bars, for better presentation, and add labels to the x and y axes.

```python
#---------------------------------------------------------------------
# Part 1: Create a figure and axis using plt.subplots
fig, ax = plt.subplots( )
#---------------------------------------------------------------------

# Part 2: Plot, modify, plot again, include text, etc.
ax.bar( means.index, means['Freshweight (g)'], width=0.5, yerr=stds['Freshweight (g)'], color='red' )
ax.plot(means.index, means['Freshweight (g)'], marker='o', color='black', markersize=15 )

ax.set_xlabel( 'Treatment', fontsize=14 )
ax.set_ylabel( 'Freshweight $[g]$', fontsize=14 )
fig.suptitle( 'Lettuce freshweight', fontsize=16 )
#---------------------------------------------------------------------
# Part 3: Show, save to file, and close
plt.show( fig ) 
plt.close( fig ) 
#---------------------------------------------------------------------
```

If you feel a little bit uneasy because the error bars don't have the upper and lower horizontal lines, caps, you can know that it is a feature of several styles, like `seaborn`. If you want to have more control over your error bars, you can use `ax.errorbar()`, instead of including it directly in the call to `ax.bar()`. 

Here an example of the use of errorbars:

```python
#---------------------------------------------------------------------
# Part 1: Create a figure and axis using plt.subplots
fig, ax = plt.subplots( )
#---------------------------------------------------------------------

# Part 2: Plot, modify, plot again, include text, etc.
ax.bar( means.index, means['Freshweight (g)'], width=0.5, color='red', linewidth=3, edgecolor='black' )
ax.errorbar( means.index, means['Freshweight (g)'], stds['Freshweight (g)'], color='black', capsize=7, markeredgewidth=2 )
ax.plot(means.index, means['Freshweight (g)'], marker='d', color='black', markersize=15 )

x_labels = [ 'Treatment 1', 'Treatment 2 - \n A very long treatment', 'Control' ]
ax.set_xticklabels( x_labels, fontsize=16, rotation=30, ha='right' )

ax.set_xlabel( 'Treatment', fontsize=14 )
ax.set_ylabel( 'Freshweight $[g]$', fontsize=14 )
fig.suptitle( 'Lettuce freshweight', fontsize=16 )
#---------------------------------------------------------------------
# Part 3: Show, save to file, and close
plt.show( fig ) 
plt.close( fig ) 
#---------------------------------------------------------------------
```

### Air temperature and relative humidity: Twin axes

So far, the diagrams have had a single x and a single y-axes. However, it is commonly needed to plot two series of data using different scales. In these cases, it is common to link one series to the left scale, and the other to the right.

We can achieve this in `matplotlib` using the `twinx()` function.

We will build up on the temperature diagram from before, to include the relative humidity on the right-hand axis.

The diagram was like this:

```python
#---------------------------------------------------------------------
# Part 1: Create a figure and axis using plt.subplots
fig, ax = plt.subplots()
#---------------------------------------------------------------------

# Part 2: Plot, modify, plot again, include text, etc.
ax.plot( one_day_only.index, one_day_only['Temperature'] )

#Format axes
ax.xaxis.set_major_locator( mdates.HourLocator( byhour=range(0,24,5) ) )
ax.xaxis.set_major_formatter( mdates.DateFormatter('%d/%b \n %H:%M') ) 

#---------------------------------------------------------------------
# Part 3: Show, save to file, and close
plt.show( fig ) 
plt.close( fig ) 
#---------------------------------------------------------------------
```

If a second series of data is included, it will be drawn on the left-side axis (the only one, by that matter), resulting then in a problem with scale. In this case, temperature and relative humidity have values in different ranges, making it difficult to see the actual values on the curve:

```python
#---------------------------------------------------------------------
# Part 1: Create a figure and axis using plt.subplots
fig, ax = plt.subplots()
#---------------------------------------------------------------------

# Part 2: Plot, modify, plot again, include text, etc.
ax.plot( one_day_only.index, one_day_only['Temperature'], color='red' )
ax.plot( one_day_only.index, one_day_only['Relative Humidity'], color='blue' )

#Format axes
ax.xaxis.set_major_locator( mdates.HourLocator( byhour=range(0,24,6) ) )
ax.xaxis.set_major_formatter( mdates.DateFormatter('%H:%M') ) 

#---------------------------------------------------------------------
# Part 3: Show, save to file, and close
plt.show( fig ) 
plt.close( fig ) 
#---------------------------------------------------------------------
```

To include a second axis, we take the existing one (`ax`) and create a second one as its 'twin'.

We then use this second axis to plot, as we do with the first one. The data won't mix. 

The use of `twinx` should be understood as _they are twins, they share the same 'x'_.

```python
ax_right = ax.twinx()
```

```python
#---------------------------------------------------------------------
# Part 1: Create a figure and axis using plt.subplots
fig, ax = plt.subplots()
ax_right = ax.twinx()
#---------------------------------------------------------------------

# Part 2: Plot, modify, plot again, include text, etc.
ax.plot( one_day_only.index, one_day_only['Temperature'] )

ax_right.plot( one_day_only.index, one_day_only['Relative Humidity'] )

#Format axes
ax.xaxis.set_major_locator( mdates.HourLocator( byhour=range(0,24,6) ) )
ax.xaxis.set_major_formatter( mdates.DateFormatter('%H:%M') ) 

#---------------------------------------------------------------------
# Part 3: Show, save to file, and close
fig.tight_layout()
plt.show( fig ) 
plt.close( fig ) 
#---------------------------------------------------------------------
```

The function `fig.tight_layout()` often helps to give a slightly better distribution, particularly when there are different axes and legends.

There remains only giving some format to the graph:

```python
#---------------------------------------------------------------------
# Part 1: Create a figure and axis using plt.subplots
fig, ax = plt.subplots( figsize=(9,6) )
ax_right = ax.twinx()
#---------------------------------------------------------------------

# Part 2: Plot, modify, plot again, include text, etc.
ax.plot( one_day_only.index, one_day_only['Temperature'], color='red', label='Air temperature' )
ax.plot( one_day_only.index[0], [0], color='blue', label='Relative humidity' ) # Trick: doesn't print anything, only used to get the legend

ax_right.plot( one_day_only.index, one_day_only['Relative Humidity'], color='blue', label='Relative humidity' )

#Format axes
ax.xaxis.set_major_locator( mdates.HourLocator( byhour=range(0,24,6) ) )
ax.xaxis.set_major_formatter( mdates.DateFormatter('%H:%M') ) 

ax.set_xlabel( 'Time', fontsize=14 )
ax.set_ylabel( 'Temperature $[^\circ C]$', fontsize=14 )
ax_right.set_ylabel( 'Relative humidity $[\%]$', fontsize=14 )

ax.set_ylim( [10,32] )
ax_right.set_ylim( [20,100] )

ax_right.grid( False )

ax.legend( loc='center left' )
#ax_right.legend( loc='top right' )

fig.suptitle( 'Air conditions at the agroforestry plot\n ' + str( start.date() ), fontsize=18 )

#---------------------------------------------------------------------
# Part 3: Show, save to file, and close
fig.tight_layout( rect=[0,0,1,0.99] )
plt.show( fig ) 
plt.close( fig ) 
#---------------------------------------------------------------------
```

Last note: If you have a very heavy plot, you might want to plot it to a file directly, without rendering it to the screen. This can be particularly useful in two situations:
