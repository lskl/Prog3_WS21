```python
import pandas as pd
df = pd.json_normalize(jsondata['data'])
```

```python
df = df.rename( columns=jsondata['header'] )
```

```python
df = df.set_index( pd.DatetimeIndex( df['Datetime'] ) )
df.head()
```

```python
df.columns
```

```python
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
```

```python
fig, ax = plt.subplots( figsize=(15,7) )

ax2 = ax.twinx()

ax.plot( df.index, df['Lufttemp_o - C'], color='red' )
ax2.plot( df.index, df['PAR_o - mV'], color='black', alpha=0.5 )
ax2.fill_between( df.index, df['PAR_o - mV'], color='gold', alpha=0.5 )

ax.xaxis.set_major_locator( mdates.HourLocator( byhour=range(0,24,12) ) )
ax.xaxis.set_major_formatter( mdates.DateFormatter('%d-%b\n%H:%M') ) 



plt.show()
```

```python
fig, ax = plt.subplots( figsize=(15,7) )

ax2 = ax.twinx()

ax.plot( df.index, df['Theta_10 - %'], color='blue', 
            marker='^', markevery=30, label='Soil water content at 10cm' )
ax.plot( df.index, df['Theta_25 - %'], color='darkblue', 
            marker='o', markevery=30, label='Soil water content at 25cm' )
ax.plot( df.index, df['Theta_55 - %'], color='navy', 
            marker='>', markevery=30, label='Soil water content at 55cm' )
ax.plot( df.index, df['Theta_115 - %'], color='black', 
            marker='v', markevery=30, label='Soil water content at 115cm' )

ax2.plot( df.index, df['Rain#@15m - mm'], color='cyan', label='Rain' )


ax.xaxis.set_major_locator( mdates.HourLocator( byhour=range(0,24,12) ) )
ax.xaxis.set_major_formatter( mdates.DateFormatter('%d-%b\n%H:%M') ) 

ax.legend()
ax2.legend()

plt.show()
```
